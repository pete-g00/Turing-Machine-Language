```python
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
```

```python
files = pd.read_csv("results.csv")
values = files.values
```

```python
values[(values=="Strongly Disagree") | (values=="Strongly disagree")] = -1
values[values=="Disagree"] = -0.5
values[values=="Agree"] = 0.5
values[(values=="Strongly Agree") | (values=="Strongly agree")] = 1
```

```python
values[values=="No"] = 0
values[values=="Maybe"] = 0.5
values[values=="Yes"] = 1
```

```python
tml_evaluation = values[:, 12:16].astype(np.float16)
```

```python
fig, ax = plt.subplots()
fig.set_size_inches(12,6)
plt.violinplot(tml_evaluation)
plt.title("TML Evaluation")
plt.xlabel("Criterion")
plt.xticks([1, 2, 3, 4])
plt.ylabel("Agreement")
```

```python
comparison_evaluation = values[:, 18:20].astype(np.float16)
```

```python
fig, ax = plt.subplots()
fig.set_size_inches(12,6)
plt.violinplot(comparison_evaluation)
plt.title("TML vs TM")
plt.xlabel("Criterion")
plt.xticks([1, 2])
plt.ylabel("Agreement")
plt.ylim(bottom=-1)
```

```python
use_tml = values[:, 20]
```

```python
fig, ax = plt.subplots()
fig.set_size_inches(12,6)
plt.hist(use_tml)
plt.title("Use of TML")
plt.xlabel("Agreement")
plt.ylabel("Count")
plt.xticks([0, 0.5, 1])
```

```python
fig, ax = plt.subplots()
fig.set_size_inches(12,6)
plt.violinplot(website_evaluation)
plt.title("Product Evaluation")
plt.xlabel("Criterion")
plt.ylabel("Agreement")
```
